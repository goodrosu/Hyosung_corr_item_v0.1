```python
import pandas as pd
import pandas_datareader.data as web
from datetime import datetime
import numpy as np
import random
import time
import matplotlib.pyplot as plt

# 구글계정 입력 
google_username = "OOOO@gmail.com"
google_password = "OOOO"

# 구글트렌드 검색기간 입력
time_term = '2013-03-01 2017-03-31'
start = datetime(2013, 3, 1)
end   = datetime(2017, 3, 31)
```

```python
# 키워드 A 입력
keyword='스판덱스'
from pytrends.request import TrendReq
pytrend = TrendReq(google_username, google_password, custom_useragent='Python Trends')
pytrend.build_payload(kw_list=[keyword], timeframe=time_term)
key_a = pytrend.interest_over_time()
key_a = key_a[keyword].resample('Q', how='mean') # 구글 트렌드 분기별 평균
key_a.head()
```

```python
# 키워드 C 입력
keyword='아라미드'
from pytrends.request import TrendReq
pytrend = TrendReq(google_username, google_password, custom_useragent='Python Trends')
pytrend.build_payload(kw_list=[keyword], timeframe=time_term)
key_c = pytrend.interest_over_time()
key_c = key_c[keyword].resample('Q', how='mean') # 구글 트렌드 분기별 평균
key_c.head()
```

```python
# 키워드 D 입력
keyword='폴리케톤'
from pytrends.request import TrendReq
pytrend = TrendReq(google_username, google_password, custom_useragent='Python Trends')
pytrend.build_payload(kw_list=[keyword], timeframe=time_term)
key_d = pytrend.interest_over_time()
key_d = key_d[keyword].resample('Q', how='mean') # 구글 트렌드 분기별 평균 
key_d.head()
```

```python
# 종목별 주가데이터(시가총액) & 재무데이터(매출액, 영업이익, 당기순이익, 매출액, 자본총계, ROE 등)
hyosung = pd.read_csv('C:\\Users\\user\hyosung.crm', sep=',', header=None)
hyosung.columns = ['temp', '매출채권', '판관비용', '영업이익', '설비투자', '이자비용', '부채비율', '감가상각비', '이익잉여금', '현금창출능력', '부채증가분', '재고자산', '시가총액', '당기순이익', '순운전자본', '급여비용', '자기자본이익률', '매출', '매출원가', '매출총이익', '자본총계', '총부채']
hyosung.head()
```

```python
# 데이터 병합
hyosung.index = hyosung['temp']
key_a.index = hyosung['temp']
key_b.index = hyosung['temp']
key_c.index = hyosung['temp']
key_d.index = hyosung['temp']
data = pd.concat([hyosung,
                  key_a, key_b, key_c, key_d], axis=1)
data2 = data.drop('temp', axis=1)
data3 = data2.drop('총부채', axis=1)
data3.head()
```

```python
# 결측값 체크
data3.isnull().sum() 
data3.isnull().sum().value_counts()

# 로그수익률 계산
rets = np.log(data3 / data3.shift(1))
rets.head()
```

```python
# 상관계수 계산
df_corr = rets.corr()
df_corr.head()
```

```python
# 결과물
def corr_bottom(code, bottom_cnt=-10):
     return df_corr[df_corr.index == code].T\
                 .sort_values(by=code, ascending=False)\
                 .ix[-bottom_cnt:]                  
corr_bottom('시가총액', 30)
```
